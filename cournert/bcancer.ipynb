```python
x_train, x_test, y_train, y_test = train_test_split(x,y, stratify=y, test_size=0.2, random_state=2)
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_predicion = model.predict(x_train)
traning_accuracy = accuracy_score(x_train_predicion, y_train)
print("training prediction accuracy is :", traning_accuracy)
```

```python
x_test_predicion = model.predict(x_test)
test_accuracy = accuracy_score(x_test_predicion, y_test)
print("test prediction accuracy is :", test_accuracy)
```
