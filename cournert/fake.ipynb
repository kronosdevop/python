```python
from faker import Faker
fake = Faker()
```

Timeseries

```python
def fake_timeseries_rows(num=1, seed=0):
    """
    """
    fake.seed_instance(seed)
    
    time_gen = fake.time_series(start_date= f"-{num}d", end_date="now", precision=3600)
    
    output = [
        {
            "datetime" : next(time_gen),
            "name" :  fake.name(),
            "email" : fake.free_email(),
            "country" : fake.country(),
            "department": fake.job(),
            "D-o-B" : fake.passport_dob(),
            "Passport" : fake.passport_number(),
            "Phone-num" :  fake.phone_number(),
        }
        for x in range(num)
    ]
    return output
```

```python
df = pd.DataFrame(fake_timeseries_rows(1000, seed=3))
df
```
